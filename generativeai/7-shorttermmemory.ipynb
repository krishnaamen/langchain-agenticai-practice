### Agent with short term memory

```python
import os
from dotenv import load_dotenv
load_dotenv()
openai_key = os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
```

```python
from groq.types import model
agent = create_agent(
    model = "gpt-5",
    checkpointer=InMemorySaver()
)
```

```python
config = {"configurable":{"thread_id": "test1"}}
```

```python
from langchain_core.messages import HumanMessage, SystemMessage
response = agent.invoke({"messages":[HumanMessage(content="Hi My name is Krishna")]},config=config)
response
```

```python
response = agent.invoke({"messages":[HumanMessage(content="What is my name?")]},config=config)
response
```
