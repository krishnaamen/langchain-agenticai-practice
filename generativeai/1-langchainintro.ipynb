```python
import os 
from langchain_groq import ChatGroq


model = ChatGroq(model="qwen/qwen3-32b")
response = model.invoke("who is the primeminister of Nepal")
print(response.content)
```

```python
for chunk in model.stream("why parrots have colorful feathers?"):
    print(chunk.text,end="|",flush=True)
```

```python
responses = model.batch(
    ["Why parrots have colorful feather?",
    "How do airplanes fly",
    "How to land AI Engineering job"]
)
for response in responses:
    print(response)
```
