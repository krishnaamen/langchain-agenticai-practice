```python
import os
os.environ["TAVILY_API_KEY"] = os.getenv("TAVILY_API_KEY")

from langchain_groq import ChatGroq
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = ChatGroq(model="qwen/qwen3-32b")
```

```python
from langchain_tavily import TavilySearch

tool1 = TavilySearch(
    max_results=5,
    topic="general"

)
#tool.invoke("who is Krishna Prasad Khanal, a software developer, based on copenhagen denmark who worked in crm nordics as a developer")
```

```python

from langchain.tools import tool

@tool("calculator", description="Performs arithmetic calculations. Use this for any math problems.")
def calc(expression: str) -> str:
    """Evaluate mathematical expressions."""
    return str(eval(expression))
```

```python
from langchain.agents import create_agent

agent = create_agent(
    model = model,
    tools=[tool1,calc]
)
```

```python
user_input = "What is the present condition of worldcup Argentina vs Austria and then calculate 5 + 5"
```

```python

for step in agent.stream(
    {
        "messages": user_input},
        stream_mode= "values"
    
):
    step["messages"][-1].pretty_print()
```
